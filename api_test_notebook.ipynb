```python
# Library for the Zoho API
import requests
import json
import os
import datetime as dt
import json

class serpStackApi:
    def __init__(self):
        self.base_url = "http://api.serpstack.com/search"
        self.access_key = "cee82cc5cabe5928b1077465dc4c805f"

    def search(self, query, location, language):
        # Construct the API request URL
        url = f"{self.base_url}?access_key={self.access_key}&query={query}&auto_location=0&gl={location}&hl={language}"

        try:
            # Make the HTTP GET request
            response = requests.get(url)

            # Check if the request was successful (status code 200)
            if response.status_code == 200:
                # Return the JSON response
                return response.json()
            else:
                # Print an error message if the request was not successful
                print(f"Error: {response.status_code} - {response.text}")
                return None

        except Exception as e:
            # Handle any exceptions that may occur during the request
            print(f"Error: {e}")
            return None

def elt_serpstack(input_keyword, input_country, input_language):

    result = serpStackApi().search(query = input_keyword, location = input_country, language = input_language)

    if result:
        # Print the result
        return result
    else:
        print("Failed to retrieve search results.")
        return None

# API parameters
json_result = elt_serpstack('mcdonalds', 'us', 'en')

# Specify the file path
file_path = r'C:\Users\idbs9\Downloads\serp.json'

# Save the dictionary to a JSON file
with open(file_path, 'w', encoding='utf-8') as json_file:
    json.dump(json_result, json_file, ensure_ascii=False, default=str)

print(f"Result saved to {file_path}")

# Open and read the content of the JSON file
with open(file_path, 'r', encoding='utf-8') as json_file:
    loaded_data = json.load(json_file)
```

```python
import csv

# Function to Iterate throught the API and the fields
def save_results_to_csv(columns, data, csv_file_path, result_type):
    """
    Save search results data to a CSV file.

    Parameters:
        - columns: The JSON structure to be iterated
        - data: The loaded JSON data containing search results.
        - csv_file_path: The path to the CSV file where the data will be saved.
        - result_type: Type of result ('organic', 'paid', or other types you might add).
    """

    columns = columns_mapping.get(result_type, {})
    if not columns:
        print(f"Invalid result type: {result_type}")
        return

    with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
        # Create a CSV writer object
        csv_writer = csv.writer(csv_file)

        # Write the header row
        csv_writer.writerow(columns.keys())

        # Iterate through each result
        for result in data.get(result_type, []):
            # Extract values based on columns mapping
            row_data = [result.get(column_key, '') for column_key in columns.values()]

            # Write a row for each result
            csv_writer.writerow(row_data)

    print(f"{result_type.capitalize()} results saved to {csv_file_path}")


# Define columns mapping for each result type
columns_mapping = {
    'organic_results': {
        'Position': 'position',
        'Title': 'title',
        'URL': 'url',
        'Displayed URL': 'displayed_url',
        'Snippet': 'snippet',
        'Cached Page URL': 'cached_page_url',
        'Related Pages URL': 'related_pages_url'
    },
    'ads': {
        'Position': 'position',
        'Block Position': 'block_position',
        'Title': 'title',
        'URL': 'url',
        'Tracking URL': 'tracking_url',
        'Displayed URL': 'displayed_url',
        'Description': 'description',
        'Sitelinks': 'sitelinks'
    },
    'request': { 
        'Success': 'success',
        'Processed Timestamp': 'processed_timestamp',
        'Search URL': 'search_url',
        'Total Time Taken': 'total_time_taken'
    },
    'search_parameters': { 
        'Engine': 'engine',
        'Query': 'query',
        'Type': 'type',
        'Device': 'device',
        'Google Domain': 'google_domain',
        'HL': 'hl',
        'GL': 'gl',
        'Page': 'page',
        'Num': 'num'
   },
    'search_information': { 
        'Total Results': 'total_results',
        'Time Taken Displayed': 'time_taken_displayed',
        'Did You Mean': 'did_you_mean',
        'Showing Result For': 'showing_results_for',
        'Query Displayed': 'query_displayed',
        'Detected Location': 'detected_location',
        'No Results for Original Query': 'no_results_for_original_query'
   }
}

# Organic Results
organic_results_csv_path = r'C:\Users\idbs9\Downloads\organic_results_csv.csv'
save_results_to_csv(columns_mapping, loaded_data, organic_results_csv_path, 'organic_results')

# Paid Results
paid_results_csv_path = r'C:\Users\idbs9\Downloads\paid_results_csv.csv'
save_results_to_csv(columns_mapping, loaded_data, paid_results_csv_path, 'ads')
```
